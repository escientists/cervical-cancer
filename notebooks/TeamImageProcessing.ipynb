```python
from IPython.display import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import io, color, feature
from skimage.filters import rank
from skimage.color import rgb2grey
from skimage import data
image = mpimg.imread("/data/cervicalCancer/train/Type_1/104.jpg")
igray = rgb2grey(image)
plt.imshow(igray,cmap = plt.get_cmap('gray'))

plt.show(figsize=(100,100))
    
    
```

```python
from skimage.morphology import disk
ent = rank.entropy(igray, disk(5))

plt.imshow(ent,cmap = plt.get_cmap('gray'))
plt.show()
```
